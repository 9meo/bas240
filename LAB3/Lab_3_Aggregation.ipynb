# Data Aggregation

- A process of summarizing or combining multiple pieces of data into a single, unified dataset or value.
- By applying operations/functions to **group** and **condense** data into a more digestible form.
- Operation/function: **summing**, **averaging**, **counting**, or finding the **minimum** or **maximum**

- The groupby function in pandas is a powerful tool that allows you to group data based on one or more columns and then apply aggregate functions to each group.

How groupby Works:
1. Splitting the Data: The data is split into groups based on the values in one or more columns. Each unique value (or combination of values) forms a group.

2. Applying a Function: Once the data is grouped, you can apply aggregate functions (like sum, mean, max, min, count, etc.) to each group independently.

3. Combining the Results: The results of the function applied to each group are combined into a new DataFrame or Series, where each row represents a group and the values represent the result of the aggregation.

```python
grouped = df.groupby('column_name').aggregate_function()
```

- df.groupby('column_name'): This creates a GroupBy object, where the data is grouped by the unique values in column_name.
- aggregate_function(): This is the function applied to each group, such as sum(), mean(), or count().

````python

df.groupby('Product')
````

````python
df.groupby('Product').count()
````

````python
#Total quantity sold of each product

df.groupby('Product')['Quantity'].count()
````

````python
# Average sale per day
df.groupby('Date')['Sales'].mean()
````
