# Pivot Table

Pivot table is a tool used in data analysis that helps organizing
and summarizing large amounts of data in a simple, easy-to-read
format.
- A pivot table is a powerful data analysis tool that helps to summarize and reorganize data in a structured format.
- Take large datasets and "pivot" the data.
- Transforming rows into columns and columns into rows to provide a more comprehensive view of the data.

### Example 1: Total Sales by Product and Region
Objective: Summarize the total sales for each product across different regions.

```python
pivot_table_sales = df.pivot_table(values='Sales', index='Product', columns='Region', aggfunc='sum')
print(pivot_table_sales)
```

Explanation:

- values='Sales': We want to aggregate the Sales data.
- index='Product': The rows of the pivot table will be the different products.
- columns='Region': The columns of the pivot table will represent the different regions.
- aggfunc='sum': We’re summing the sales figures to get total sales for each product in each region.



````python
df.pivot_table(values='Sales', index='Product', columns='Region', aggfunc='sum')
````

Interpretation:

- Jeans had the highest sales in the North region with a total of 410.
- Sneakers did well in the South and East regions, with 710 and 365 in sales, respectively.
- The North region shows the highest sales for Jeans and T-shirts, indicating strong market demand in that region for these products.
- Sneakers dominate in the South region with the highest recorded sales across all products and regions.
- The East and West regions have moderate sales for all products, with no extreme highs or lows.

# Unpivot Table in Pandas
Unpivoting (also known as "melting") is the process of transforming data from a wide format, where each column represents a variable, to a long format, where each row represents a single observation. This is particularly useful when you need to prepare data for analysis or visualization, as many tools and functions work better with long-format data.

**Example**: Unpivoting a Table
Let's start with a wide-format table, similar to the pivot table you provided, and then unpivot it to a long format.

Original Wide-Format Data:
We'll use a similar structure to the pivot table in your image:


````python
import pandas as pd

data = {
    'Product': ['Jeans', 'Sneakers', 'T-shirt'],
    'East': [385, 365, 265],
    'North': [410, None, 440],
    'South': [365, 710, None],
    'West': [345, 325, 230]
}

df_wide = pd.DataFrame(data)
print("Original Wide-Format Table:")
df_wide
````

This table shows the sales of different products in different regions in a wide format.

**Unpivoting the Table**:

We can unpivot this table using the **pd.melt()** function in pandas:


````python
df_long = pd.melt(df_wide, id_vars=['Product'], var_name='Region', value_name='Sales')
print("Unpivoted (Long-Format) Table:")
df_long
````

**Explanation**:

- id_vars=['Product']: This specifies the columns that should remain fixed (i.e., not unpivoted).
- var_name='Region': This is the name of the new column that will contain the - names of the old column headers (East, North, South, West).
- value_name='Sales': This is the name of the new column that will contain the - values from the original table.

**Interpretation of the Unpivoted Table**:

Each row in the unpivoted table represents a single observation of product sales in a specific region.

- The Region column now contains the values "East," "North," "South," and "West," which were originally the column headers in the wide-format table.
- The Sales column contains the corresponding sales figures.
- This long-format table is more suited for certain types of data analysis, especially when using functions that require a single variable in one column.

**Why Use Unpivoting?**

Unpivoting is useful when:

- You need to prepare data for visualization tools that expect long-format data.
- You're performing certain types of analyses, such as statistical modeling, where data must be in a long format.
- You're combining datasets that are in different formats and need to standardize them.
